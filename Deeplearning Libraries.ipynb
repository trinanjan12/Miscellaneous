## Matplotlib

There are thousands of libraries in Python, and Matplotlib is one of the most powerful tools for data visualization in Python.
Matplotlib tries to make easy things easy and hard things possible. You can generate plots, histograms, power spectra, bar charts, errorcharts, scatterplots, etc., with just a few lines of code.

```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
import numpy as np

# Prepare the data
x = np.linspace(0, 10, 100)

# Plot the data
plt.plot(x, x, label='linear')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

## Seaborn
Seaborn builds on top of Matplotlib and introduces additional plot types. It also makes your traditional Matplotlib plots look a bit prettier.



```python
# example taken from here
# https://seaborn.pydata.org/examples/heatmap_annotation.html

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Load the example flights dataset and convert to long-form
flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax)
```

## Keras 

is where most of the people start their deep learning journey with. Keras is very simple and easy to read and interpret. It offeres fucntions which are very simple to use and users don't have to think much before using it. But problem starts coming when you want to dig deeper and want to change something very specific to the job.
keras is basicially build with tensorflow as backend()


What is the relationship between tensorflow with keras?

One is a high level library. That’s Keras. Keras is high level, meaning it’s much easier to code with than authoring TF natively.

One is a Framework. That’s TensorFlow.

Keras sits on top of TF… and a few other frameworks… CNTK… etc. That means you can’t use Keras without an “engine.”

You can use TF alone because it is an “engine.”

TF was authored by Deep Brain, a subsidiary of Google.

Keras was authored by François Chollet while he was an intern at Google.

Authoring core TF isn’t the most fun thing to do. Authoring Keras is much easier.

Google realized the community liked TF the engine but not TF the programming aspect. TF 2.0 is about to be released and Google wisely did two things. They embedded Keras into TF 2.0 and the new release.
That means you can rapid prototype your deep learning models in Keras while maintaining granular control of the underlying knobs and switches in TF.


## TensorFlow

is Google's open source AI framework for machine learning and high performance numerical computation. TensorFlow is a Python library that invokes C++ to construct and execute dataflow graphs

Coding with TensorFlow directly may take some time because it comes down to almost low level of programming(near to machine) .But tensorflow could be faster. And it offers not only building models but also deploying it so that it could be used in real time. BDW keras in their backend uses tensorflow. 

what's a backend:

It is not possible to only use Keras without using a backend, such as Tensorflow, because Keras is only an extension for making it easier to read and write machine learning programs. All the actual calculations needed to create models are not implemented in Keras, which is why you need to use a backend library for anything to work.
